```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('review-Alaska_reduced_v2.csv')
```

```python
df.rename(columns={'/gmap_id' : 'Business ID', '/name' : 'Name', '/rating' : 'Rating', '/text' : 'Review', '/time' : 'Time', '/user_id' : 'User ID'}, inplace = True)
df
                   
                   
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.count()
```

```python
df['Readable Time'] = pd.to_datetime(df['Time'], unit='ms')
df
```

```python
df.head()
```

```python
df['Review'] = df['Review'].astype(str)
df['totalwords'] = df['Review'].str.count('')+1
df
```

```python
df = df.drop(df[df.totalwords < 5].index)
df
```

```python
df = df.drop(df[df.totalwords < 6].index)
df
```

```python
data_dup = df.duplicated().any()
print(data_dup)
```
